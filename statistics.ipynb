```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
rv = stats.expon(scale=3.0)
```

```python
r = rv.rvs(size=1000) + stats.norm(scale=0.01).rvs(size=1000)
```

```python
def plot(rv, r):
    x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 100)
    fig, ax = plt.subplots(1, 1)
    ax.plot(x, rv.pdf(x), 'r-', lw=5, alpha=0.6, label='expon pdf')
    ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
    ax.legend(loc='best', frameon=False)
    plt.show();
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
plt.show();
```

```python
plot(stats.expon(scale=r.mean()), r)
```

```python
stats.kstest(r, stats.expon(scale=r.mean()).cdf)
```
